Import Libraries

```python
import pandas as pd
from datetime import datetime, date
import os
```

Read data from today's silver file

```python
silver_dir = "/home/tigureis/DNC_engenharia_de_dados/DncInsight_Solution/airflow/data/silver"
today_date = datetime.today().strftime('%Y-%m-%d')
silver_file = os.path.join(silver_dir, f"silver_{today_date}.csv")
    
if not os.path.exists(silver_file):
    raise FileNotFoundError(f"No silver file found for today: {today_date}")
        
df = pd.read_csv(silver_file)


df.head(10)
```

```python
df
```

Calculate the *'subscription_time'* for active members.
Set as 0 for the inactive ones

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
today = pd.to_datetime(datetime.today().strftime('%Y-%m-%d'))
df['subscription_time'] = 0
df.loc[df['subscription_status'] == True, 'subscription_time'] = ((today - df['signup_date']).dt.days / 365.25)
df
```

Set a colum with the *'age_groups'*

```python
df["age_group"] = df["age"].apply(lambda age: f"{(age // 10) * 10}-{(age // 10) * 10 + 9}")
df
```

Agg the data by *agg_groups*

```python

agg_age_df = df.groupby(df['age_group']).agg(
        total_users=('id', 'count'),
        active_users=('subscription_status', lambda x: (x == True).sum()),
        inactive_users=('subscription_status', lambda x: (x == False).sum()),
        avg_sub_time_for_active_users=('subscription_time', lambda x: x[df.loc[x.index, 'subscription_status'] == True].mean())
    ).reset_index()
agg_age_df
```

Agg the data by *'subscription_status'*

```python
agg_status_df = df.groupby(df['subscription_status']).agg(
        total_users=('id', 'count'),
        avg_age=('age','mean'),
        avg_subscription_time=('subscription_time', 'mean')
    ).reset_index()
agg_status_df
```

Creat **.parquet**s files separated by *'age_group'*

```python
output_dir = "/home/tigureis/DNC_engenharia_de_dados/DncInsight_Solution/airflow/data/gold/age_group"
os.makedirs(output_dir, exist_ok=True)

for age_group, group in df.groupby("age_group"):
    file_path = os.path.join(output_dir, f"age_group_{age_group}.parquet")
    group.to_parquet(file_path, index=False)
    print(f"Saved {file_path}")

print("Parquet files saved successfully.")
```
